### Copiar el ID

link: https://docs.google.com/spreadsheets/d/```=> desde aqui se copia el ID de la hoja de calculo de google sheet 1mSM41UfQHQjo2msBlHvQn8mYyYOm57rv5eHsUnlLvXM```/edit?usp=sharing

### Guardamos en una variable el ID.

```sheet_id =``` 1mSM41UfQHQjo2msBlHvQn8mYyYOm57rv5eHsUnlLvXM




### Siguiente paso

Se quita lo que esta al final del link ```edit?usp=sharing``` Y se agrega al final del link ```gviz/tq?tqx=out:csv&sheet```

### Dentro del f'' String se cooloca:
    1.Se coloca el link original.
    2.Se reemplaza el ID y se coloca la variable donde se guardo dentro de {}.
    3.Se agrega al final ```gviz/tq?tqx=out:csv&sheet``` lo que conecta con el Sheet

````python
sheet_id = '1mSM41UfQHQjo2msBlHvQn8mYyYOm57rv5eHsUnlLvXM'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
````

````python
datos_co2 = pd.read_csv(url)
datos_co2
````
