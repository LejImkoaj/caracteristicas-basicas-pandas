# Leyendo el Archivo Excel

```read_excel``` se puede usar para leer un archivo de Excel en formato xlsx. Sin embargo, esta función es capaz de leer archivos en otros formatos, como: ```xls, xlsm, xlsb, odf, ods y odt```.

pandas.read_excel ```( io , ​​sheet_name=0 ,*, header =0 , names=Ninguno , index_col=Ninguno , usecols=Ninguno , dtype=Ninguno , engine=Ninguno , converters=Ninguno , true_values=Ninguno , false_values=Ninguno , skiprows=Ninguno , nrows=Ninguno , na_values=Ninguno , keep_default_na=True , na_filter=True , verbose=False , parse_dates=False , date_parser=<sin_predeterminado> , date_format=Ninguno , miles=Ninguno , decimal='.' , comment=Ninguno , skipfooter=0 , storage_options=Ninguno , dtype_backend=<sin_predeterminado> , engine_kwargs=Ninguno )```
