# Normalizando archivo JSON anidado

```python
df_normalizado = pd.json_normalize(datos_pacientes2['Pacientes'])
df_normalizado
```

```python
#Importando el módulo JSON
import json

#Leyendo el archivo json usando el módulo Python JSON
with open('pacientes_2.json','r', encoding='utf-8') as f: # Abrir el archivo JSON
    datos = json.loads(f.read()) # Cargando los datos del archivo JSON

# Si 'Pacientes' es una lista
df = pd.json_normalize(
    datos, 
    record_path='Pacientes', # 
    meta=['Investigación', 'Año'], # Metadatos a incluir
    errors='ignore'  # Esto ayuda con columnas que podrían faltar
)

df

```

# Exportar JSON

```python
df.to_json('historico_pacientes_norma.json')
```
